```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

obs_df = pd.read_csv("observations.csv")
info_df = pd.read_csv("species_info.csv")
all_df = obs_df.merge(info_df, how="left", left_on="scientific_name", right_on="scientific_name")
#df1.merge(df2, how='left', left_on='col1', right_on='x')
```

```python
all_df
```

```python
sns.kdeplot(all_df.observations, hue=all_df.conservation_status, shade=True)
```

```python
hue_order = ["Yellowstone National Park","Yosemite National Park", "Bryce National Park", "Great Smoky Mountains National Park"]
sns.catplot(x="conservation_status", y="observations", hue="park_name", kind="bar", data=all_df, height=5, aspect=2, hue_order=hue_order)
```

```python
### Left off trying to make bar chart of the different animal categories, and how many there are on each conservation_status
df_test = all_df.groupby("category").conservation_status.value_counts()
sns.catplot(x="category", y="conservation_status", kind="bar", data=df_test)
```
