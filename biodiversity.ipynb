The goal of this analysis is to provide some insight that these National Parks and other interested parties can use to make decisions. In depth knowledge on the biodiversity among these parks will be able to drive future direction and help to learn on the mistakes and successes of the past.

### Data Manipulation:

-Below I cleaned up the data by removing duplicates, filling null values, and creating different types of dataframes

-Those 5 dataframes included:
        
1. An Observations dataframe with expanded/pivoted values to show how many observations per park as individual columns
    - Not used in any visualizations
2. A Species dataframe that has null values and duplicates dropped
    - Used in all of the "Figure 1" visualizations
3. A Species dataframe that groups animal categories and counts the unique animals for each category
    - Not used in any visualizations, but used for description
4. A Species dataframe that groups animal categories with a subset of their conservation status, and how many unique animals fall into those categories
    - Not used in any visualizations, but used for description
5. A dataframe that merges the first two dataframes, the Observation and the Species to create a master dataframe
    - Used in all of the "Figure 2" visualizations


````python
print("MANIPULATING THE OBSERVATIONS DATAFRAME:\n")

print("-dropping duplicates")
observations_fix = obs_df.drop_duplicates()

print("-grouping based on scientific_name and park_name")
observations_groupby = observations_fix.groupby(["scientific_name", "park_name"]).observations.mean().reset_index()

print("-creating a pivoted chart to expand observations per park")
observations_pivoted = observations_groupby.pivot(index="scientific_name",
                                                  columns="park_name",
                                                  values="observations").reset_index()
observations_pivoted.columns = ["scientific_name", "bryce", "smoky", "yellowstone", "yosemite"]


print("\n")
print(observations_pivoted.head())
print("\n\n")



print("MANIPULATING THE SPECIES DATAFRAME:\n")

print("-dropping duplicates")
species_fix = info_df.drop_duplicates(subset=["scientific_name", "conservation_status"])

print("-fill null values")
species_fillna = species_fix.fillna("None")

print("-drop null values")
species_dropna = species_fix.dropna()

print("-create a separate dataframe that counts categories of animals (see below for head)")
species_groupby_cat = species_fillna.groupby("category").scientific_name.count().reset_index()
species_groupby_cat.columns = ["category", "count"]

print("-create a separate dataframe that counts categories of animals with a conservation status(see below for head)")
species_groupby_cat_dropna = species_dropna.groupby("category").scientific_name.count().reset_index()
species_groupby_cat_dropna.columns = ["category", "count"]
species_groupby_cat["wstatus"] = species_groupby_cat_dropna["count"]
species_groupby_cat["percentage"] = species_groupby_cat.apply(lambda x: x["wstatus"]/x["count"]*100, axis=1)


print("-create a separate dataframe that counts categories of animals by their conservation status (see below for head)")
species_groupby_status = species_fillna.groupby(["category", "conservation_status"]).scientific_name.count().reset_index()
species_groupby_status.columns = ["category", "conservation_status", "count"]

print("\n")
print(species_fillna.head())
print("\n")

print("Group by Category:")
print(species_groupby_cat.head())

print("\nGroup by Category only with Status")
print(species_groupby_cat_dropna.head())

print("\nGroup by Conservation Status:")
print(species_groupby_status.head())

print("\n\n")

print("MERGING ALL INTO ONE DATAFRAME\n")

print("-merge the pivoted obeservations df with the species_fillna df")
merged_df = observations_pivoted.merge(species_fillna, how="left")

print("\n")
print(merged_df)
print("\n")
````

### Data Visualization

-There are two sets of Figures for visualizations, and seven total visualizations to analyze

- Figure 1.1 - 1.3 are bar charts that display counts of unique animals in the data defined by different subsets:
    - Species with a Conservation status
    - Species counts broken up by conservation status
    - All species in the data (no conservation status needed)


- Figure 2.1 - 2.4 are violin plots that display the observations by conservation status of the four parks from the data:
    - Bryce National Park
    - Greak Smoky Mountains National Park
    - Yellowstone National Park
    - Yosemite National Park

````python
def count_cat_bar():
    sns.set_style("darkgrid")
    sns.set_context("talk")
    fig, ax = plt.subplots(figsize=(20,10))
    ax = sns.countplot(data=species_fillna[species_fillna.conservation_status != "None"], x="category", \
                      order=species_fillna[species_fillna.conservation_status != "None"].category.value_counts().index)

    for p in ax.patches:
        ax.text(x = p.get_x() + p.get_width() / 2, y = p.get_height() + 0.1,
                s = '{:.0f}'.format(p.get_height()), ha = 'center', color = '#4f4e4e')
    
    plt.title("Total Count of Unique Species with a Conservation Status per Category", fontsize=25)
    plt.xlabel("Category", labelpad=15, fontsize=20)
    plt.ylabel("Count", labelpad=15, fontsize=20)
    
    plt.figtext(0.1, 0.015, "Figure 1.1", fontsize=15, fontstyle="italic")
    
    plt.savefig("Count_Categories_wStatus_Bar")

    plt.show()
````

````python
def count_cat_by_convs_bar():
    sns.set_style("darkgrid")
    sns.set_context("talk")
    fig, ax = plt.subplots(figsize=(20,10))
    ax = sns.countplot(data=species_fillna[species_fillna.conservation_status != "None"], x="category", \
                       hue="conservation_status", order=species_fillna[species_fillna.conservation_status != "None"].category.value_counts().index)

### Big thanks to aleks-andra-sasha on this one, had some troubles with the labeling
    for p in ax.patches:
        if float(p.get_height()) > 0:
            height = p.get_height()
        else:
            height = .000001
        ax.text(x = p.get_x() + p.get_width() / 2, y = height + 0.1,
                s = '{:.0f}'.format(height), ha = 'center', color = '#4f4e4e')
    
    plt.legend(loc=1)
    plt.title("Total Count of Unique Species by Conservation Status per Category", fontsize=25)
    plt.xlabel("Species Category", labelpad=15, fontsize=20)
    plt.ylabel("Count", labelpad=15, fontsize=20)
    
    plt.figtext(0.1, 0.015, "Figure 1.2", fontsize=15, fontstyle="italic")
    
    plt.savefig("Count_Cat_by_Status_Bar")

    plt.show()
````

````python
def count_cat_no_cons_bar():
    sns.set_style("darkgrid")
    sns.set_context("talk")
    fig, ax = plt.subplots(figsize=(20,10))
    ax = sns.countplot(data=species_fillna, x="category", order=species_fillna.category.value_counts().index)

    for p in ax.patches:
        ax.text(x = p.get_x() + p.get_width() / 2, y = p.get_height() + 0.1,
                s = '{:.0f}'.format(p.get_height()), ha = 'center', color = '#4f4e4e')
    
    plt.title("Total Count of Unique Species without a Conservation Status per Category", fontsize=25)
    plt.xlabel("Category", labelpad=15, fontsize=20)
    plt.ylabel("Count", labelpad=15, fontsize=20)
    
    plt.figtext(0.1, 0.015, "Figure 1.3", fontsize=15, fontstyle="italic")
    
    plt.savefig("Count_Categories_Bar")

    plt.show()
````

````python
count_cat_bar()
count_cat_by_convs_bar()
count_cat_no_cons_bar()
````

````python
def obs_by_park_and_cons_violin(parkdataname, parkname, fignum):
    sns.set_style("darkgrid")
    sns.set_context("talk")
    fig, ax = plt.subplots(figsize=(20,10))
    
    ax = sns.violinplot(data=merged_df, x="conservation_status", y=parkdataname)
    
    plt.title("Observations at {} National Park by Conservation Status".format(parkname), fontsize=25)
    plt.xlabel("Conservation Status", labelpad=15, fontsize=20)
    plt.ylabel("Observations", labelpad=15, fontsize=20)
    
    plt.figtext(0.1, 0.015, "Figure {}".format(fignum), fontstyle="italic", fontsize=15)
    
    plt.savefig("Obs_by_Status_{}".format(parkdataname.title()))

    plt.show()
````
